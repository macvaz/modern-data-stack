```python
from pyspark.sql.functions import lit
from functools import reduce
```

```python
parquet_df = spark.read.parquet("/home/iceberg/data/*.parquet")
parquet_df.printSchema()
```

```python
names = [f"serie_{i}" for i in range(1500)]
PAGE_SIZE = 10
```

```python
names_page = names[0:PAGE_SIZE]
    %time page1_df = reduce(lambda df, serie: df.withColumn(serie, lit(False)), names_page, parquet_df)
page1_df.printSchema()
```

```python
%time page1_df.write.format("parquet").mode("overwrite").save("/tmp/page1")
```

```python
names_page = names[PAGE_SIZE:2*PAGE_SIZE]
%time page2_df = reduce(lambda df, serie: df.withColumn(serie, lit(False)), names_page, parquet_df)
page2_df.printSchema()
```

```python
%time page1_df.write.format("parquet").mode("overwrite").save("/tmp/page1")
```

```python
names_page = names[2*PAGE_SIZE:3*PAGE_SIZE]
%time page3_df = reduce(lambda df, serie: df.withColumn(serie, lit(False)), names_page, parquet_df)
page3_df.printSchema()
```

```python
page3_df.write.format("parquet").mode("overwrite").save("/tmp/page3")
```
